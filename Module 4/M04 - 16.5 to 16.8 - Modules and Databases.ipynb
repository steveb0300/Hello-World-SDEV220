```python
# Steve Baker
# Module 4 - 16.5 to 16.8 - Programming Assignment - Modules and Databases

import sqlite3
import pandas as pd
from sqlalchemy.engine import result 
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String, select, text

# Step 16.4: Create the SQLite database and books table
conn = sqlite3.connect('books.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS books (
    title TEXT,
    author TEXT,
    year INTEGER
)
''')

conn.commit()

# Step 16.5: Read books2.csv and insert its data into the books table
books_df = pd.read_csv('books.csv')
books_df.to_sql('books', conn, if_exists='append', index=False)

# Step 16.6: Select and print the title column from the books table in alphabetical order
cursor.execute('SELECT title FROM books ORDER BY title')
titles = cursor.fetchall()
print("Titles in alphabetical order:")
for title in titles:
    print(title[0])

# Step 16.7: Select and print all columns from the books table in order of publication
cursor.execute('SELECT * FROM books ORDER BY year')
books = cursor.fetchall()
print("\nBooks in order of publication:")
for book in books:
    print(book)

conn.close()

# Step 16.8: Use the sqlalchemy module to connect to the sqlite3 database and select the title column in alphabetical order
engine = create_engine('sqlite:///books.db')

# Initialize metadata
metadata = MetaData()

# Define the table structure (assuming the table is named 'books')
books = Table('books', metadata,
              Column('id', Integer, primary_key=True),
              Column('title', String),
              Column('author', String),
              Column('pub_year', Integer))

# Reflect the table from the database
metadata.reflect(bind=engine)

# Check if the table is correctly reflected
if 'books' in metadata.tables:
    books = metadata.tables['books']
else:
    print("Table 'books' not found in the database.")

with engine.connect() as connection:
    sql = text('SELECT title FROM books ORDER BY title')  
    result = connection.execute(sql).fetchall()
    print("\nTitles in alphabetical order (using SQLAlchemy):")
    for row in result:
        print(row[0])
```
